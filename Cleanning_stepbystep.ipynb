```python
import pandas as pd
import numpy as np
df = pd.read_csv('C:/Users/ASUS/Documents/Python Scripts/Headphones/headphones_final.csv')
```

```python
#cleanning stars, by adding them in a new column named rating
stars = {r'.*star-full.*': 1, r'.*start-half.*':0.5,r'.*start-empty.*':0} #if the string havs the term x, replace it for the assigned value
df['star1'] = df.star1.replace(stars, regex = True)
df['star2'] = df.star2.replace(stars, regex = True)
df['star3'] = df.star3.replace(stars, regex = True)
df['star4'] = df.star4.replace(stars, regex = True)
df['star5'] = df.star5.replace(stars, regex = True)

df['rating'] = df.iloc[:,-5:].sum(axis =1) # sum all the values and store them in the new column rating
#df = df.drop(columns = ['star1', 'star2','star3','star4','star5']) #drop the other column

df
```

```python
df['rating'] = df.iloc[:,-5:].sum(axis =1) # sum all the values and store them in the new column rating
#df = df.drop(columns = ['star1', 'star2','star3','star4','star5']) #drop the other column
```

```python
#Due to some of the prices have two dots "1.999.000", it couldnt transform to float.  This algorithm made it.
df['Price'] = df['Price'].str.replace(".","", regex = True).astype(int)
```

```python
#extract the number from the string in the reviews
A=df['Reviews'].str.split(pat = '>',n = 1,expand = True)
B = A[1].str.split(pat = '<',n = 1,expand = True)
B[[0]]
df['Num_reviews'] = B[0]
df = df[['Title','Price','Num_reviews','rating']]
```

```python
#Covertir todo a minuscula
df['Title'] = df['Title'].str.lower()
```

```python
#Drop rows that contain
df = df[~df.Title.str.contains('limpiador|caja|cable para|Espuma|laptop|almohadilla|tablet|portatil|celular|macbook')]
df['Title'] = df['Title'].str.replace("á","a")
df['Title'] = df['Title'].str.replace("í","i")
df['Title'] = df['Title'].str.replace("é","e")
df['Title'] = df['Title'].str.replace("ó","o")
df['Title'] = df['Title'].str.replace("ú","u")
```

```python
#convert NaN in 0's
df['Num_reviews']=df['Num_reviews'].fillna(0)
```

```python
#Convert the data type of reviews into float
df['Num_reviews'] = df['Num_reviews'].astype(float)
```

```python
#Extract the brand from the title

Brands = {r'.*bose.*':'bose',
          r'.*beyerdynamic.*':'beyerdynamic',
          r'.*pioneer.*':'pioneer',
          r'.*beats.*':'beats',
          r'.*lenovo.*':'lenovo',
          r'.*marshall.*':'marshall',r'.*audio-technica.*':'audio-technica', r'.*aftershokz.*':'aftershokz',r'.*sennheiser.*':'sennheiser',
          r'.*sony.*':'sony',r'.*soundcore.*':'soundcore',r'.*logitech.*':'logitech',r'.*skullcandy.*':'skullcandy', r'.*huawei.*':'huawei',
          r'.*jbl.*':'jbl',r'.*samsung.*':'samsung',r'.*akg.*':'akg',r'.*honor.*':'honor',r'.*behringer.*':'behringer',
          r'.*haylou.*':'haylou',r'.*xiaomi.*':'xiaomi',r'.*lenovo.*':'lenovo',r'.*motorola.*':'motorola',r'.*maxell.*':'maxell',r'.*apple.*':'apple',
          r'.*oppo.*':'oppo',r'.*airpods.*':'apple',r'.*HyperX.*':'HyperX',r'.*Soundpeats.*':'Soundpeats',r'.*Astro.*':'Astro',r'.*Onikuma.*':'Onikuma',
          r'.*Corsair.*':'Corsair',r'.*Xtrike.*':'Xtrike',r'.*Razer.*':'Razer',r'.*Kotion.*':'Kotion',r'.*Genius.*':'Genius',r'.*kz.*':'kz'
         }


df['Brand'] = df.Title.replace(Brands, regex = True)
df.loc[df.Brand.str.contains('\.'),'Brand'] = 'other'

#Bose|Beyerdynamic|Pioneer|Beats|Marshall|Audio-Technica|AfterShokz|Sennheiser|Sony|Soundcore|Logitech|Skullcandy|Huawei|JBL|Samsung|AKG|Honor|Behringer|Haylou|Xiaomi|Lenovo|Maxell|Motorola|Apple|HyperX|Soundpeats|Astro|Onikuma|Corsair|Xtrike|Razer|Kotion|Genius
```

```python
designs = {r'.*in-ear.*':'in-ear',r'.*interno.*':'in-ear',r'.*earbuds.*':'in-ear'}
df['design'] = df.Title.replace(designs, regex = True)
df.loc[~df.design.str.contains('in-ear'),'design'] = 'Headphones'

Connection = {r'.*inalambrico.*':'bluetooth', r'.*bluetooth.*':'bluetooth'}
df['conn'] = df.Title.replace(Connection, regex = True)
df.loc[~df.conn.str.contains('bluetooth'),'conn'] = 'wire'

Usage = {r'.*gamer.*':'gamer'}
df['Use'] = df.Title.replace(Usage, regex = True)
df.loc[~df.Use.str.contains('gamer'),'Use'] = 'No-gamer'
```

```python
#Drop those rows without brand, put the brand in capital letter, order the table
df = df[~df.Brand.str.contains(' ')]
df['Brand'] = df['Brand'].str.upper()
df1 = df[['Title','Brand','design','conn','Use','Num_reviews','rating','Price']]
```

```python
countries = {'BOSE':'USA','BEYERDYNAMIC':'Germany','PIONEER':'Japan',
             'BEATS':'USA','MARSHALL':'UK','AUDIO-TECHNICA':'Japan',
             'AFTERSHOKZ':'USA','SENNHEISER':'Germany','SONY':'Japan',
             'SOUNDCORE':'China','LOGITECH':'Swiss','SKULLCANDY':'USA',
             'HUAWEI':'China','JBL':'USA','SAMSUNG':'Korea','AKG':'Austria',
             'HONOR':'China','BEHRINGER':'Germany','HAYLOU':'China',
             'XIAOMI':'China','LENOVO':'China','MAXELL':'Japan','MOTOROLA':'USA',
             'APPLE':'USA','HYPERX':'USA','SOUNDPEATS':'China','ASTRO':'USA',
             'ONIKUMA':'Japan','CORSAIR':'USA','XTRIKE':'China','RAZER':'USA',
             'KOTION':'China','GENIUS':'China','KZ':'China','OTHER':'China'
}

df1['country'] = df.Brand.replace(countries, regex = True)
```

```python
#adding the segment of market

Segment  = {'BOSE':'professional','BEYERDYNAMIC':'professional','PIONEER':'professional',
             'BEATS':'high-end','MARSHALL':'high-end','AUDIO-TECHNICA':'professional',
             'AFTERSHOKZ':'high-end','SENNHEISER':'high-end','SONY':'high-end',
             'SOUNDCORE':'high-end','LOGITECH':'high-end','SKULLCANDY':'high-end',
             'HUAWEI':'Mobil accessories','JBL':'high-end','SAMSUNG':'Mobil accessories','AKG':'high-end',
             'HONOR':'Mobil accessories','BEHRINGER':'professional','HAYLOU':'Mobil accessories',
             'XIAOMI':'Mobil accessories','LENOVO':'Mobil accessories','MAXELL':'Mobil accessories','MOTOROLA':'Mobil accessories',
             'APPLE':'Mobil accessories','HYPERX':'high-end','SOUNDPEATS':'Mobil accessories','ASTRO':'high-end',
             'ONIKUMA':'high-end','CORSAIR':'high-end','XTRIKE':'gamer accessories','RAZER':'high-end',
             'KOTION':'gamer accessories','GENIUS':'Mobil accessories','KZ':'gamer accessories','OTHER':'gamer accessories','OPPO':'Mobil accessories'
}
df1['segment'] = df1.Brand.replace(Segment, regex = True)
```
